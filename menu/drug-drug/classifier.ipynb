```python
import os
import csv
from dotenv import load_dotenv
import textwrap
import google.generativeai as genai
from IPython.display import Markdown

# Define file path
file_path = 'C:\\Users\\khand\\Desktop\\med_AI\\text.csv'

# Function to convert text to Markdown
def to_markdown(text):
    text = text.replace('•', '  *')
    text = text.replace('/n', '')
    return Markdown(textwrap.indent(text, '> ', predicate=lambda _: True))

def read_raw_file(file_path):
    try:
        with open(file_path, 'r') as file:
            content = file.read()
            return content
    except FileNotFoundError:
        return "Error: File not found"
    except Exception as e:
        return f"Error reading file: {str(e)}"

# Read and print the raw content
raw_content = read_raw_file(file_path)

# Load environment variables
load_dotenv()
google_api_key = os.getenv('GEMINI_API_KEY')

# Configure Gemini API
genai.configure(api_key=google_api_key)

# Define system prompt
system_prompt_ddc = {"text": """You are a medical AI Agent. Classify on the basis:
                1. Major: Highly clinically significant. Avoid combinations; the risk of the interaction outweighs the benefit.
                2. Moderate: Moderately clinically significant. Usually avoid combinations; use it only under special circumstances.
                3. Minor: Minimally clinically significant. Minimize risk; assess risk and consider an alternative drug, take steps to circumvent the interaction risk and/or institute a monitoring plan 1 word."""}

user_prompt = raw_content

# Function to call Gemini API
def message_gemini(user_prompt):
    gemini = genai.GenerativeModel(
        'gemini-1.5-flash',
        system_instruction=system_prompt_ddc,
    )
    response = gemini.generate_content(user_prompt)
    return response.text

    # Call the function and print the response
    result = message_gemini(user_prompt)
    print(result)
```

```python
message_gemini(user_prompt)
```
