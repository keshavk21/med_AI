```python
import os
from dotenv import load_dotenv
import json
import textwrap
import google.generativeai as genai
from IPython.display import Markdown

class DrugInteractionAnalyzer:
    def __init__(self, api_key):
        """Initialize the analyzer with API key and configure Gemini."""
        self.configure_api(api_key)
        self.system_prompt = {
            "text": "You are a medical AI agent. Respond the reaction between the drugs to human when taken together in 100 words"
        }
        
    @staticmethod
    def configure_api(api_key):
        """Configure the Gemini API with the provided key."""
        genai.configure(api_key=api_key)
    
    @staticmethod
    def to_markdown(text):
        """Convert text to markdown format."""
        text = text.replace('•', '  *')
        return Markdown(textwrap.indent(text, '> ', predicate=lambda _: True))
    
    @staticmethod
    def load_json_file(file_path):
        """Load and parse JSON file."""
        try:
            with open(file_path, "r") as file:
                return json.load(file)
        except FileNotFoundError:
            raise FileNotFoundError(f"Could not find file at {file_path}")
        except json.JSONDecodeError:
            raise ValueError("Invalid JSON format in the file")
    
    def analyze_interactions(self, drugs):
        """Analyze drug interactions using Gemini API."""
        if not drugs:
            return {"error": "No drugs provided"}
        
        user_prompt = ", ".join(drugs)
        
        try:
            gemini = genai.GenerativeModel(
                'gemini-1.5-flash',
                system_instruction=self.system_prompt
            )
            response = gemini.generate_content(user_prompt)
            return response.text
        except Exception as e:
            return {"error": f"API Error: {str(e)}"}

def main():
    # Define file path
    file_path = "C:\\Users\\khand\\Desktop\\med_AI\\text.json"
    
    # Load environment variables
    load_dotenv()
    google_api_key = os.getenv('GEMINI_API_KEY')
    
    try:
        # Initialize analyzer
        analyzer = DrugInteractionAnalyzer(google_api_key)
        
        # Load and process data
        data = analyzer.load_json_file(file_path)
        drugs = data.get("drugs", [])
        
        # Get and print results
        result = analyzer.analyze_interactions(drugs)
        print(result)
        
    except Exception as e:
        print(f"Error: {str(e)}")

if __name__ == "__main__":
    main()
```
