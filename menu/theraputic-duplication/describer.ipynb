```python
import os
from dotenv import load_dotenv
import json
import textwrap
import google.generativeai as genai
from IPython.display import Markdown

# Define file path
file_path = "C:\\Users\\khand\\Desktop\\med_AI\\test.json"

# Function to convert text to Markdown
def to_markdown(text):
    text = text.replace('•', '  *')
    text = text.replace('/n','')
    return Markdown(textwrap.indent(text, '> ', predicate=lambda _: True))

# Function to load JSON file
def load_json_file(file_path):
    with open(file_path, "r") as file:
        data = json.load(file)  # Load JSON data
    return data

# Load environment variables
load_dotenv()
google_api_key = os.getenv('GEMINI_API_KEY')

# Configure Gemini API
genai.configure(api_key=google_api_key)

# Define system prompt
system_prompt_tdd ={"text":"""You are a medical AI Agent.Tell is there any theraputic duplication in the list in 100 words"""}

# Load data from JSON file
data = load_json_file(file_path)

# Extract drugs list from JSON
drugs = data.get("drugs", [])
if not drugs:
    print({"error": "No drugs provided"})
else:
    user_prompt = ", ".join(drugs)

    # Function to call Gemini API
    def message_gemini(user_prompt):
        gemini = genai.GenerativeModel(
            'gemini-1.5-flash',
            system_instruction=system_prompt_tdd,
        )
        response = gemini.generate_content(user_prompt)
        return response.parts

    # # Call the function and print the response
    result = message_gemini(user_prompt)
    print(result)
```
